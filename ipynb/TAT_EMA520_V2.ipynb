```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

def filter_data(df, days):
    num_of_trade_days = df['TradeDayIdx'].max() + 1
    return df.loc[df['TradeDayIdx'] >=  (num_of_trade_days - days)].copy()

def rank_trading_days(df):
    profit_per_day = df.groupby('EntryDayOfWeek')['ProfitLossAfterSlippage'].sum() * 100
    return profit_per_day.sort_values(ascending=False)

def rank_trading_times(df):
    df['EntryHourMinute'] = df['EntryTime'].dt.strftime('%H:%M')
    profit_per_time = df.groupby('EntryHourMinute')['ProfitLossAfterSlippage'].sum() * 100
    return profit_per_time.sort_values(ascending=False)

def calculate_pivot_table(df):
    df['EntryHour'] = df['EntryTime'].dt.hour
    df['EntryMinute'] = df['EntryTime'].dt.minute
    return df.pivot_table(index='EntryHour', columns='EntryMinute', values='ProfitLossAfterSlippage', aggfunc='sum', fill_value=0)

def plot_heatmap(pivot_table, title):
    # Create a colormap where negative values are red and positive values are green, and 0 values are black
    colors = sns.diverging_palette(10, 130, n=256).as_hex()
    colors = ['black'] + colors
    cmap = ListedColormap(colors)

    # Multiply the pivot table values by 100
    pivot_table_scaled = pivot_table * 100

    plt.figure(figsize=(10, 10))
    sns.heatmap(pivot_table_scaled, cmap=cmap, center=0, cbar=True, annot=True, linewidths=.5, fmt=".2f")
    plt.title(title, fontsize=14)
    plt.xlabel('Minute', fontsize=12)
    plt.ylabel('Hour', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()


# Load data from the CSV file with datetime parsing
df = pd.read_csv("Trades.csv", parse_dates=['EntryTime'])

# Add new columns for date of entry and zero-based trade day index
df['EntryDate'] = df['EntryTime'].dt.date
df['TradeDayIdx'] = df.groupby(['EntryDate']).ngroup()

days_list = [90, 60, 45, 30, 20, 15, 10, 5]

for days in days_list:
    df_filtered = filter_data(df, days)
    print(f"\nBest times to enter a trade for the last {days} days:")
    print(rank_trading_times(df_filtered))
    pivot_table = calculate_pivot_table(df_filtered)
    plot_heatmap(pivot_table, f'Profit and Loss per Entry Time (Hour and Minute) - Last {days} Days')

print("\nBest times to enter a trade for all data:")
print(rank_trading_times(df))
pivot_table = calculate_pivot_table(df)
plot_heatmap(pivot_table, 'Profit and Loss per Entry Time (Hour and Minute) - All Data')

# Add a new column for the day of the week
df['EntryDayOfWeek'] = df['EntryTime'].dt.day_name()

# Calculate and print the profit per day of the week
print("\nProfit per day of the week:")
print(rank_trading_days(df))
```
